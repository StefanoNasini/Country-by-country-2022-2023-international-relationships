```python
import requests
from bs4 import BeautifulSoup

# URL of the target page
url = "https://www.equaldex.com/equality-index"

# Make an HTTP GET request to fetch the page content
response = requests.get(url)
html_content = response.content

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
den = x.find_all('tr')[2]
ctrys = x.find_all('tr')[1:]
```

```python
cname = (lambda x:x[x.find(".")+2:])(den.find('b').text)
```

```python
def parse_ctry(x):
    cname, eq, lg, pb = (lambda x: tuple(map(lambda y: y.text, x)))(x.find_all('td'))
    cname = (lambda x:x[x.find(".")+2:])(cname)
    if cname == "Iceland":
        return cname, '92', '98', '87'
    if eq == 'â\x80\x94':
        eq = "N/A"
    if lg == 'â\x80\x94':
        lg = 'N/A'
    if pb == 'â\x80\x94':
        pb = 'N/A'
    return cname, eq, lg, pb
```

```python
import pandas as pd
df = pd.DataFrame()


for x in ctrys:
    cname, eq, lg, pb = (parse_ctry(x))
    df.loc[cname, "Equality Index"] = eq
    df.loc[cname, "Legal Index"] = lg
    df.loc[cname, "Public Opinion Index"] = pb
    
```

```python
df.to_csv(os.path.join(root, "raw_data", "environmental_lgbt", 'world_equality_index.csv'))
```

```python
def get_dict_from_link(link):
    # URL of the target page
    url = "https://www.equaldex.com"+link

    # Make an HTTP GET request to fetch the page content
    response = requests.get(url)
    html_content = response.content

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')
    
    ips = soup.find('dl').find_all('div', class_='info_pairw')
    
    data = dict()
    for ip in ips:
        label, value = ip.find('dt', class_= "info_label").text, ip.find('strong').text
        data[label] = value
    return data
```

```python
import pandas as pd
df = pd.DataFrame()

for link in links:
    cname = link[8:]
    print(f"loading {cname}")
    data = get_dict_from_link(link)
    for k, v in data.items():
        df.loc[cname, k] = v
```

```python
df.to_csv(os.path.join(root, "raw_data", "environmental_lgbt", "lgbt_legal_status.csv"))
```
